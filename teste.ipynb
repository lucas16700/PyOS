```python
import re
from rich import print

TOKEN_REGEX = re.compile(
    r'''\s*(
        [-+]?\d+            | # números
        "(?:\\.|[^"])*"     | # strings
        [()]                | # parênteses
        [^\s()]+              # símbolos
    )''',
    re.VERBOSE
)

def tokenize(text):
    return TOKEN_REGEX.findall(text)

def parse(tokens):
    def parse_expr():
        token = tokens.pop(0)

        if token == '(':
            lst = []
            while tokens[0] != ')':
                lst.append(parse_expr())
            tokens.pop(0)  # remove ')'
            return lst

        elif token.startswith('"'):
            return token[1:-1]  # string sem aspas

        elif token.isdigit():
            return int(token)

        else:
            return token  # símbolo

    ast = []
    while tokens:
        ast.append(parse_expr())
    return ast
```

```python
teste="""(program
    (target bloco)
    (def hover (a b) 
        (change color (255 255 255))
        return 0
    )
    (event hover set hover)
)
(meta
(version 0)
(budleid (10 2 3))
(name "teste de  'ss a' config")
)
"""
print(parse(tokenize(teste)))
```
